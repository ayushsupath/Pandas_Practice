```python
import numpy as np
import pandas as pd
```

```python
#Finding Missing Data
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df.isna()
```

```python
df.isna().any()
```

```python
#Removing Missing Data

df
```

```python
df.dropna()
```

```python
df.dropna(thresh=1)
```

```python
#Filling the Missing dataabs
df
```

```python
df.fillna(0)
```
