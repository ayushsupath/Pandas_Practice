```python
#Creating a series in pandas

import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
mylist = [10,20,30]
my_arr = np.array([10,20,30])
d = {'a':10, 'b':20, 'c':30}
```

```python
pd.Series(mylist)
```

```python
pd.Series(mylist, index=labels)
```

```python
pd.Series(my_arr)
```

```python
pd.Series(d)
```
